```python
def period(string):
    
    #makes coincidence array
    n=len(string)
    coincidence=[]
    sig_coincidence=[]
    
    i=0
    while i<n:
        count=0
        
        j=0
        while j<n:
            if (string[j]==string[(j+i) % n]): 
                count+=1
            j+=1             
        if (count*100/n>8): coincidence.append([i,count*100/n])
        i+=1
  
    #finds gcd of positions in coincidence array

    n=len(coincidence)
    start=coincidence[0]
    divisor=start[0]
    i=0
    while i<n:
        #picks up a position coincidence pair
        pair=coincidence[i]
        #picks up position only
        number=pair[0]
        divisor=math.gcd(divisor,number)
        i+=1
    
    return divisor
    
    
```
